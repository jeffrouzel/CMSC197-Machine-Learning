<h1> Visualizing the impact of different coefficients </h1>

```python
from scipy.optimize import curve_fit

def objective(x, a, b):
    return a * x + b

popt, _ = curve_fit(objective, df["TV"], df["Sales"])
a, b = popt
print('Best line fit: y = %.5f * TV + %.5f' % (a, b))

x_line = np.arange(min(df["TV"]), max(df["TV"]), 1)
y_line = objective(x_line, a, b)
y_line2 = objective(x_line, 0.03, b)
y_line3 = objective(x_line, 0.03, 6)
y_line4 = objective(x_line, 0, b)
y_line5 = objective(x_line, a, 0)

plt.subplots(figsize=(11, 9))
plt.scatter(df["TV"], df["Sales"])
plt.xlabel("TV")
plt.ylabel("Sales")
plt.plot(x_line, y_line, '--', color='red', label = 'y = '+str(np.round(a,2))+'TV + '+str(np.round(b,2)))
plt.plot(x_line, y_line2, '--', color='orange',label = 'y = 0.03TV + '+str(np.round(b,2)))
plt.plot(x_line, y_line3, '--', color='pink',label = 'y = 0.03TV + 6')
plt.plot(x_line, y_line4, '--', color='green',label = 'y = 0*TV +'+str(np.round(b,2))+' (slope = 0)')
plt.plot(x_line, y_line5, '--', color='blue',label = 'y = '+str(np.round(a,2))+'TV + 0 (bias = 0)')
plt.legend()
plt.title("Sales vs TV")
```

<h2> Split dataset into train and test set </h2>

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 42)
```

```python
print("x_train size:", x_train.shape)
print("y_train size:", y_train.shape)
print("\nx_test size:", x_test.shape)
print("y_test size:", y_test.shape)
```

```python
#compute for the mean
y_train_mean = np.mean(y_train)
y_train_mean
x_train_mean = np.mean(x_train)
x_train_mean

#compute for the beta
beta_1 = np.sum((x_train-x_train_mean)*(y_train-y_train_mean))/np.sum(np.square(x_train-x_train_mean))
beta_0 = y_train_mean-beta_1*x_train_mean
print(beta_1,beta_0)
```

```python
#saving the step by step computation into a dataframe
ols_res = pd.DataFrame()
ols_res['x'] = x_test
ols_res['ŷ'] = y_test
ols_res['β0'] = beta_0
ols_res['β1'] = beta_1
```

<h2> Predict y values (sales) using the estimated coefficients </h2>

```python
#y = B0 + B1x
y_predicted = x_test*beta_1 + beta_0

#save the predicted values into a dataframe
ols_res['ŷ = β1x + β0'] = y_predicted
```

<h2> Evaluate the model using MSE </h2>

Evaluating on the test set

```python
def MSE(y_hat, y):
    return np.sum(np.square(y_hat-y))/len(y)
```

```python
MSE(y_predicted,y_test)
```

```python
#saving the error computations into a dataframe
ols_res['y-ŷ'] = y_predicted-y_test
ols_res['e²=(y-ŷ)²'] = np.square(ols_res['y-ŷ'])
print(ols_res.head(5))
```

```python
y_predicted_train = x_train*beta_1 + beta_0
print(MSE(y_predicted_train, y_train))
```

<h1> Using SKLEARN </h1>

```python
### Load the data
df = pd.read_csv("Advertising.csv", index_col = 0)
print("Size of the data:", df.shape)
df.head()
```

```python
x = df.drop("Sales", axis = 1)
y = df["Sales"]

print(f"==x (predictors)==\nSize: {x.shape}\n{x.head()}\n Data Type: {type(x.head())} ")
print(f"\n==y (target)==\nSize: {y.shape}\n{y.head()}\n{type(y.head())}")
```

```python
### Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 42)

print("x_train size:", x_train.shape)
print("y_train size:", y_train.shape)
print("\nx_test size:", x_test.shape)
print("y_test size:", y_test.shape)
```

```python
### Initialize dataframe that will store the results
df_results = pd.DataFrame(columns = ["Predictor/s", "R2", "MAE", "MSE", "RMSE", "Cross-Val Mean"])
```

<h2> Building the Model </h2>

```python
### Make a function for Linear Regression with default values
def linreg_model(xtrain, xtest):
    ### Initialize algorithm
    linreg = LinearRegression()
    
    ### Fit the data
    linreg.fit(xtrain, y_train)
    print(linreg.coef_)
    
    ### Evaluate the model
    y_pred = linreg.predict(xtest)
    
    print("R2:", r2_score(y_pred, y_test))
    print("MAE:", mean_absolute_error(y_pred, y_test))
    print("MSE:", mean_squared_error(y_pred, y_test))
    print("RMSE:", np.sqrt(mean_squared_error(y_pred, y_test)))
    
    f, ax = plt.subplots(figsize=(11, 9))
    plt.scatter(y_pred, y_test)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Actual vs Predicted")
    
    return {"R2": r2_score(y_pred, y_test) * 100, "MAE": mean_absolute_error(y_pred, y_test), 
            "MSE": mean_squared_error(y_pred, y_test), "RMSE": np.sqrt(mean_squared_error(y_pred, y_test))}
```

<h3> Using only TV as predictor </h3>

```python
### Predict and get results
linreg_TV_results = linreg_model(x_train["TV"].values.reshape(-1,1), x_test["TV"].values.reshape(-1,1))

### Cross Validation
cv_score = cross_val_score(LinearRegression(), x["TV"].values.reshape(-1, 1), y, cv=10)

print("Cross-Val Results:", cv_score)
print("Cross-Val Mean:", cv_score.mean())

### Compile validation results
linreg_TV_results.update({"Predictor/s":"TV", "Cross-Val Mean": cv_score.mean() * 100})

### Add the results to the dataframe
df_results = df_results.append(linreg_TV_results, ignore_index=True)
```

<h3> Using all predictors </h3>
The results of this model that uses all of the predictor variable will be our basis on the next models.

```python
### Predict and get results
linreg_all_results = linreg_model(x_train, x_test)

### Cross Validation
cv_score = cross_val_score(LinearRegression(), x, y, cv=10)

print("Cross-Val Results:", cv_score)
print("Cross-Val Mean:", cv_score.mean())

### Compile validation results
linreg_all_results.update({"Predictor/s":"All", "Cross-Val Mean": cv_score.mean() * 100})

### Add the results to the dataframe
df_results = df_results.append(linreg_all_results, ignore_index=True)
```

#### Using *TV* and *Radio* as predictors
The top 2 predictor variables based on the EDA above are *TV* and *Radio*, so let's see if the combination of these variables makes a better model.

```python
linreg_TVR_results = linreg_model(x_train[["TV", "Radio"]], x_test[["TV", "Radio"]])

### Cross Validation
cv_score = cross_val_score(LinearRegression(), x[["TV", "Radio"]], y, cv=10)

print("Cross-Val Results:", cv_score)
print("Cross-Val Mean:", cv_score.mean())

### Compile validation results
linreg_TVR_results.update({"Predictor/s":"TV & Radio", "Cross-Val Mean": cv_score.mean() * 100})

### Add the results to the dataframe
df_results = df_results.append(linreg_TVR_results, ignore_index=True)
```

This model is slightly better than the model that uses all of the predictor variables.
